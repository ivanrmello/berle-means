```python
import numpy as np                                                         
import scipy as sp                                                         
import scipy.stats
```

```python
aves = [45, 48, 70, 65, 66, 54, 52, 52, 59, 68, 49, 44, 47, 60, 68]
```

```python
conf_int = scipy.stats.norm.interval(0.90, loc=np.mean(aves), scale=4/np.sqrt(len(aves)))
print("intervalo de confiança: ")
print(conf_int)
```
