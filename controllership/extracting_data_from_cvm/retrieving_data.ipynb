```python
import requests, zipfile, io #for reading zip files straight from CVM website
import pandas as pd #dataframes
from os import walk #reading directories and files
from glob import glob #reading directories and files
```

```python
#Creating a list of links to request the file one by one from CVM's website
years_list = []
for x in range(1,2021-2010+1):
  years_list.append(2010+x)
```

```python
#This part of the code loops through each year, using it on url to download .zip
##then the code unpacks all .csv files from each .zip
for year in years_list:
  r = requests.get('http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/FRE/DADOS/fre_cia_aberta_' + str(year) + '.zip')
  z = zipfile.ZipFile(io.BytesIO(r.content))
  z.extractall("/content/drive/MyDrive/Mestrado - UFG/Disciplinas/Controladoria - Julio/Artigo/databases_fre_cvm_br")
```

When i've finished downloading all .zip files from CVM on above cell, I've manually distributed them into folders by their types, before executing next cell.

```python
#I've organized files into folders, by type.
##with this following code I am creating a list of all paths that contains .csv's by type.
list_of_type_of_files = glob("/content/drive/MyDrive/Mestrado - UFG/Disciplinas/Controladoria - Julio/Artigo/databases_fre_cvm_br/*/")
```

```python
for type_of_file in list_of_type_of_files:
  print(type_of_file.split('/')[-2])
```

```python
#Now, for every type path, i will retrieve all files 
###and import them into one single dataframe by type
dfs = {}
for type_of_file in list_of_type_of_files:
  df_name = type_of_file.split('/')[-2]
  filenames = next(walk(type_of_file), (None, None, []))[2]
  li = []
  for one_file_at_a_time in filenames:
    print("Trying to parse: " + str(type_of_file) + str(one_file_at_a_time))
    df_this_time = pd.read_csv(str(type_of_file) + str(one_file_at_a_time),sep=";", error_bad_lines=False, encoding='windows-1252')
    df_this_time['origem'] = one_file_at_a_time
    li.append(df_this_time)
  frame = pd.concat(li, axis=0, ignore_index=True)
  dfs[df_name] = frame
  
```

**ATTENTION!** There are some skipped lines due to malforming file structure, see log generated on previous cell.

```python
#Exporting processed files as .xlsx
for type_of_file in list_of_type_of_files:
  name_to_acess_df = type_of_file.split('/')[-2]
  name_to_export = type_of_file.split('/')[-2] + '.xlsx'
  dfs[name_to_acess_df].to_excel('/content/drive/MyDrive/Mestrado - UFG/Disciplinas/Controladoria - Julio/Artigo/exported_final_files/' + name_to_export, index=False, engine='xlsxwriter')
```
